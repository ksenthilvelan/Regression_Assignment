```python
import pandas as p
```

```python
dataset=p.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=p.get_dummies(dataset)
dataset
```

```python
dataset=p.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
# Input dataset

independent = dataset[['age', 'bmi', 'children','sex_female', 'sex_male','smoker_no', 'smoker_yes']]
independent
```

```python
# Output dataset

dependent = dataset[['charges']]
dependent
```

```python
# using SKLearn libraries model_selection, spliting input and output values for training-set values & test-set values

from sklearn.model_selection import train_test_split
```

```python
# providing input parameters as independent, output parameters as dependent
# giving training_set data size as 70% & test_set data size as 30% of input and output dataset

train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# instead of getting all combined values in a single dataset, based on SKLearn splitting the values as 
# Training_input = x_train, Test_input = x_test and Training_Output = y_train, Test_output = y_test

x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)

x_train # to view training - input values
```

```python
x_test # to view test - input values
```

```python
y_train # to view training - output values
```

```python
y_test # to view test - output values
```

```python
# Applying Support Vector Machine Algorithm 
# based on SKLearn preprocessing Libraries, using StandardScaler()

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

# fit_tranform() calculates x_train values & transform to evaluate mean and std deviation output values
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# based on SKLearn svm libraries, using SVR()
from sklearn.svm import SVR
regressor = SVR(kernel='rbf',C=10000)
regressor
```

```python
# substituting train_set values by providing input parameter as x_train & output parameter as y_train by using fit()

regressor.fit(x_train,y_train)
regressor
```

```python
# to view bias/intercept by using intercept_
bias = regressor.intercept_
bias
```

```python
# using n_support_ , support_ , support_vectors_ 
regressor.n_support_
regressor.support_
regressor.support_vectors_
```

```python
# Evaluating Metrics by checking model efficiency for test_set values by using predict()
y_pred=regressor.predict(x_test)
y_pred
```

```python
# Evaluating Metrics values by using r2_score

from sklearn.metrics import r2_score

# Evaluating/comparing output values (y_test) & predicted values (y_pred) 
# if result of r_score is near to 1-good Model or near to 0-bad Model

r_score = r2_score(y_test,y_pred)
r_score
```

```python
# saving the model by using pickle class/function

import pickle
filename="SupportVectorMachine_Model.sav"

# dumping the model from regressor by using pickle dump(), open() with write-binary option
pickle.dump(regressor,open(filename,'wb'))
```

```python
# load and check the Model whether it's working properly by using pickle functions-load(),open() with read-binary option
loaded_model = pickle.load(open('SupportVectorMachine_Model.sav','rb'))

# check and confirm the predicted output by inputting multiple input-values
result = loaded_model.predict([[25,27,0,1,0,1,0]])
result
```
